```python
def hello():
    for i in range(100):
        print("hello")

def hi():
    for i in range(100):
        print("hi")
```

```python
from threading import Thread
t1=Thread(target=hello,name="krishna")
t2=Thread(target=hi,name="Vishnu")
```

```python
t1.start()
t2.start()
```
